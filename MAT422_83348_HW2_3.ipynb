# **2.3.1 Joint probability distributions**

The lecture notes state that in real life, we are often interested in several random variables that are related to each other.

They go on to define joint probability as the probability of two or more events happening together. A joint probability distribution shows a probability distribution for two (or more) random variables.

The example below demonstrates some useful techniques for visually exploring the relationships between two random variables beyond simple marginal tables.

```python
n = 1000

x = np.random.randn(n)
y = np.random.randn(n)

plt.hist2d(x,y,30,vmax=10)
plt.xlabel('x')
plt.ylabel('$x^2$')
plt.show()
```

```python
sns.set()
sns.jointplot(x=x,y=y).plot_joint(sns.kdeplot)
plt.show()
```

# **2.3.2. Correlation and dependence**

The class notes state that correlations are useful because they can indicate a predictive relationship that can be exploited in practice.  When two random variables X and Y are not independent, it is frequently of interest to assess how strongly they are related to one another.

Below I present two examples.  The first simulates the real-world scenario of showing the correlation between height and weight in humans.

The second uses randomly generated values to visually demonstrate correlation between the variables on a scatter plot.

```python
# Simulated data for height (in cm) and weight (in kg)
np.random.seed(42)
heights = np.random.normal(170, 10, 100)  # Heights with mean 170 cm, std 10 cm
weights = heights * 0.5 + np.random.normal(0, 5, 100)  # Weight is roughly proportional to height with some noise

# Calculate the Pearson correlation coefficient
correlation = np.corrcoef(heights, weights)[0, 1]
print(f"Pearson correlation coefficient between height and weight: {correlation:.2f}")

# Scatter plot to visualize correlation
plt.scatter(heights, weights, alpha=0.7, edgecolor='k')
plt.title('Scatter Plot of Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
# Positive correlation
x_pos = np.linspace(1, 100, 100)
y_pos = x_pos + np.random.normal(0, 10, 100)

# Calculate the Pearson correlation coefficient
correlation = np.corrcoef(x_pos, y_pos)[0, 1]
print(f"Positive correlation: {correlation:.2f}")

# Negative correlation
x_neg = np.linspace(1, 100, 100)
y_neg = -x_neg + np.random.normal(0, 10, 100)

# Calculate the Pearson correlation coefficient
correlation = np.corrcoef(x_neg, y_neg)[0, 1]
print(f"Negative correlation: {correlation:.2f}")

# No correlation
x_none = np.linspace(1, 100, 100)
y_none = np.random.normal(50, 10, 100)

# Calculate the Pearson correlation coefficient
correlation = np.corrcoef(x_none, y_none)[0, 1]
print(f"No correlation: {correlation:.2f}")

# Plot all three examples
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Positive correlation
axs[0].scatter(x_pos, y_pos, color='g')
axs[0].set_title('Positive Correlation')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Negative correlation
axs[1].scatter(x_neg, y_neg, color='r')
axs[1].set_title('Negative Correlation')
axs[1].set_xlabel('X')

# No correlation
axs[2].scatter(x_none, y_none, color='b')
axs[2].set_title('No Correlation')
axs[2].set_xlabel('X')

plt.tight_layout()
plt.show()
```

# **2.3.3. Random samples**

The class notes state that a simple random sample is a randomly selected subset of a population and often is used in practice.

We can draw conclusions about the population mean $\mu$ using the sample mean $\bar{X} = \frac{1}{n} (X_1 +    + X_n)$.

Some of the most frequently used inferential procedures are based on properties of the sampling distribution of $\bar{X}$.

The central limit theorem (CLT) indicates that the properly normalized sum of independent random variables tends toward a normal distribution even if the original variables themselves are not normally distributed.

Below I present two examples showing the creation of random samples and a visualization using a histogram to demonstrate these concepts under the real-world scenario of comparing human heights.

```python

# Generate a random sample of 1000 data points from a normal distribution
mean_height = 170  # Mean height in cm
std_dev_height = 10  # Standard deviation in cm
sample_size = 1000

sample_heights = np.random.normal(mean_height, std_dev_height, sample_size)

# Plot a histogram of the random sample
plt.hist(sample_heights, bins=30, edgecolor='k', alpha=0.7)
plt.title("Random Sample of Heights from a Normal Distribution")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Central Limit Theorem: Simulating sample means
n_samples = 1000  # Number of samples
sample_size = 50  # Size of each sample

sample_means = [np.mean(np.random.normal(mean_height, std_dev_height, sample_size)) for _ in range(n_samples)]

# Plot the distribution of sample means
plt.hist(sample_means, bins=30, edgecolor='k', alpha=0.7)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Mean Height (cm)")
plt.ylabel("Frequency")
plt.show()
```
